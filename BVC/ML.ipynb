# Supervised learning

Supervised learning (SL) is a paradigm in machine learning where input objects (for example, a vector of predictor variables) and a desired output value (also known as human-labeled supervisory signal) train a model. The training data is processed, building a function that maps new data on expected output values. An optimal scenario will allow for the algorithm to correctly determine output values for unseen instances.

Image and speech recognition, recommendation systems, and fraud detection are all examples of how supervised learning is used.

Mainly, two types of techniques are used here:

1) Classification

2) Regression

# Unsupervised learning

Unsupervised learning is a paradigm in machine learning where, in contrast to supervised learning, algorithms learn patterns exclusively from unlabeled data. There's no predetermined target. In other words, there are no historical answers that we can learn from like we ordinarily would with supervised machine learning. 

In short, we don't supervise the model. We let it discover patterns and groupings on its own.

Some applications of unsupervised learning include natural language processing, image and video analysis, anomaly detection and customer segmentation.

Some techniques used here are:

1. Clustering

2. Market Basket analysis

3. Dimension reduction 

and so on.

# Regression

Two types of variables are involved:

1. Dependent variable - only one - Y

2. Independent variable(s) - one or more - X

__In regression, Y must be a continuous variable while X need not be.__

Types of regression models:

1. Simple regression - only one independent variable

2. Multiple regression - more than one independent variable

Each of them is of two types: Linear and non-linear

# Simple linear regression

Finding the best fit line.



Residual error = $y_i - \hat{y_i}$

$y_i$ is the actual value and $\hat{y_i}$ is the predicted value by the model (using the line).

Mean squared error



Thus, effectively a linear regression model finds a line $\hat{y} = {\beta}_0 + {\beta}_1 x$ such that MSE is minimised.

Using appropriate mathematics, we get:



$\bar{x}$ = Average value of x (Sum of all $x_i$/Number of $x_i$)

$\bar{y}$ = Average value of actual y (Sum of all $y_i$/Number of $y_i$)

# Evaluation of regression models

## 1. Evaluation metrics



Mean Absolute Error (MAE)

Mean Squared error (MSE)

Root Mean Squared error (RMSE)

Relative Absolute error (RAE)

Relative squared error (RSE)

A popular metric:

$R^2 = 1 - RSE$

Higher $R^2$ implies better model.

## 2. Evaluation approaches

Note:

Training accuracy is the percentage of correct predictions when tested using training data.

Out-of-sample accuracy is the percentage of correct predictions when tested using data on which the model was not trained.

### a. Train and test on the same dataset

Train the model on the whole dataset and test on a portion of the same dataset.

This yields high training accuracy and low out-of-sample accuracy. 

Most likely yields an overfit.

### b. Train/Test split

Split the given (known) dataset into two parts (not necessarily equal). Train it on one part and test it using the other.

Better out-of-sample accuracy.

More realistic

But one problem is that the accuracy is highly dependent on which part of the dataset was used for training and which part for testing.

This problem can be solved by __K-fold cross-validation__.
